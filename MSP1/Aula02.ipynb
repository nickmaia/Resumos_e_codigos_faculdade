## CLASSIFICAÇÃO DE MODELOS

---

##### De acordo com o ***grau de detalhamento***, temos:

---

### ***Microscópico:***
- Considera o ``sistema como um contínuo``, isto é, posição dentro do sistema é importante;
- É feito um balanço diferencial para massa, energia e quantidade de movimento;
- A característica desses modelos é que as ``variáveis dependentes são função do espaço``;
- O modelo é composto por um conjunto de equações diferenciais parciais ``(EDPs) para o processo dinâmico`` e por um conjunto de equações diferenciais ordinárias ``(EDOs) para processo invariante no tempo``;
- Podendo ser em uma, duas ou três dimensões.

### ***Macroscópico:***
- Ignora todos os detalhes internos ao sistema e, consequentemente, ``nenhum gradiente espacial é envolvido no modelo``.
- As variáveis dependentes representam valores médios sobre o volume do sistema;
- O modelo é formado por um conjunto de equações diferenciais ordinárias ``(EDOs) para o sistema dinâmico`` e por um conjunto de ``equações diferenciais algébricas para sistemas invariantes no tempo``;
- O modelo é obtido pelo balanço de massa, energia e quantidade de movimento em um volume de controle.

---

##### Para definir ***modelos de acordo com a natureza das equações***, temos:

---

### Determinístico X Estocástico

---

- Os ***modelos determinísticos*** são modelos cujas variáveis e  parâmetros possuem ``valores fixos`` para qualquer conjunto de  condições. Baseados em análise de causa e efeito;
- Os ***modelos estocásticos*** são modelos para os quais as variáveis ou os parâmetros usados para descrever as relações entre entrada e saída e  a estrutura dos elementos não são conhecidos com precisão. Contém elementos do modelo que são de natureza ``probabilística``.

---

### Linear X Não linear

---

O ***modelo linear*** é aquele que obedece ao princípio da superposição e o ***modelo não linear*** é aquele que não obedece.


---

### PRINCÍPIO DA SUPERPOSIÇÃO

---

O princípio da superposição permite que o efeito total de várias fontes independentes atuando ao mesmo tempo possa ser obtido como a soma dos efeitos individuais de cada fonte independente quando cada uma delas atua em diferentes instantes de tempo.

- ***Homogeneidade:*** se x(t) é a solução então cx(t) é também solução (em que c é uma constante);



- ***Aditividade:*** se x1 é solução e x2 também é solução então x1+x2 é solução.

### Fenomenológico X Empírico

---

#### **Modelo fenomenológico**
O modelo fenomenológico é baseado em fenômenos (transferência de massa, calor e momento); Baseado em cáculos prontos como por exemplo balanços.




#### **Modelo empírico**
O modelo empírico é baseado em dados de entrada e saída, experimentos e observações;
Mas utilizado em laboratórios.


---

### Contínuo X Discreto

---

No ***modelo contínuo*** as variáveis dependentes são definidas ao longo do ``espaço/tempo contínuo``;



No ***modelo discreto*** as variáveis são definidas somente para valores discretos de tempo e/ou espaço, ou seja, ``variam discretamente em pontos específicos do tempo`` simulado.



No ***modelo híbrido***, São modelos que possuem ``comportamento contínuo e discreto``.

## Como os modelos são obtidos?

---

### Modelagem **caixa branca**

---

- É usado o conhecimento de engenharia de processo para descrever as características físico-químicas que ocorrem no sistema no nível exigido pelo objetivo da modelagem.
- Também conhecida como ``modelagem fenomenológica``;
- Neste caso, não se usa diretamente ``os dados medidos`` pelo processo para a obtenção do modelo. Mas, ``são usados através da forma e do valor dos parâmetros do sistema``.
- mecanicista

---

### Modelagem **caixa preta**

---

- Usado quando ``não se tem conhecimento detalhado sobre os fenômenos físicoquímicos do processo``. Este conhecimento está disponível apenas através dos dados de medição e não de mecanismos subjacentes;
- Os ``dados de medição das entradas e saídas são usados para a construção do modelo do processo``. A necessidade de estimular ou excitar o sistema para obter informações úteis para a identificação do modelo é uma questão importante;
- Esse tipo de modelagem é também conhecida como ``empírica`` ou ``identificação de sistemas``.


---

### Modelagem **caixa cinza**

---

- Na maioria dos casos é usado o ``conhecimento a priori de engenharia de processos para determinar a estrutura`` e alguns dos ``valores dos parâmetros do modelo``. Em seguida, os dados medidos são usados para construir o modelo.

- Isto envolve a definição de ``mecanismos cinéticos e de transporte``, a estimativa de parâmetros, bem como a validação do modelo contra as especificações de desempenho.
