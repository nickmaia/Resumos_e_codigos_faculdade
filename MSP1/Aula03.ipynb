<h2 align="center" style="color:black">Construção de modelos</h2>

---

#### Na construção de modelos, podemos seguir o seguinte passo a passo:
- Definir as ``Variáveis dependentes fundamentais`` (V.D.F).
- Definir as ``Variáveis de estados`` (V.E).
- Definir o ``Volume de controle (V.C)``.
- ``Princípios básicos`` (PB).

---

<h3 align="center" style="color:black">Sistematização da construção de modelos
</h3>

---

### V.D.F:
 São as grandezas principais, **massa**, **energia** e **quantidade de movimento**.


---

### V.E:
São as variáveis que definem o sistema, massa específica, concentração, temperatura, pressão, vazão ...

---

### V.C:
É o volume do sistema a ser estudado.

---

### PB:
São os princípios fundamentais de conservação de massa, energia e quantidade de movimento.

<h3 align="center" style="color:gray">Hipóteses Simplificadoras</h3>

---

São considerações com embasamento teórico com finalidade de simplificar a modelagem.

#### Por exemplo um CSTR aquecido, tem as seguintes hipóteses:
1. Tanque perfeitamente misturado – propriedades físicas constantes;
2. Não há perda por evaporação;
3. Perdas de energia desprezíveis;
4. Geometria do tanque é perfeita;
5. Líquido não volátil;
6. ...
