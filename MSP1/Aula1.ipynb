### SISTEMA DE PROCESSO

Pode ser definido como a combinação de vários equipamentos com a finalidade de desempenhar uma função específica;
Cada equipamento pode promover mudanças físicas ou químicas;

### Equipamentos
Em um sistema de processos os equipamentos são:



Onde: 
- u: entradas
- y: saídas
- x: Estados
- d: Perturbações
- y = S(u,d)

---

### Classificação dos sistemas de processos

Os sistemas de processos são classificados quanto ao número de variáveis de entrada e saída, sendo eles:
#### sistemas monovariávies: (SISO – Single Input / Single Output)
São aqueles que dispões de uma variável de entrada e uma variável de saída;

#### sistemas multivariáveis: (MIMO – Multiple Input / Multiple Output)
São aqueles que dispões de múltiplas variáveis de entrada e múltiplas variáveis de saída.


---

### SIMULAÇÃO
Uma simulação é uma imitação aproximada da realidade, operação de um processo ou sistema que representa sua operação ao longo do tempo.

Os objetivos da simulação são:
- Projeto de novas unidades ( Projeto de Processos);
- Melhoria na operação de plantas existentes (Otimização de Processos);
- Controle do ponto de operação frente às perturbações (Controle de Processos).

Formas de simulação:
- Física: o próprio processo ou planta piloto – forma demorada, cara e as vezes impossível de aplicar.
- Matemática: equações matemáticas dos principais fenômenos que ocorrem no processo - quanto maior a aproximação da realidade, mais complexos serão as equações matemáticas encontradas.

---

### Modelo
O modelo é uma representação simplificada da realidade, embora distorcido e "errado", que revele alguns componentes essenciais de uma realidade complexa.
Os modelos podem ser relacionados com vários tipos alguns físicos, outros matemáticos.

#### MODELO DE PROCESSO
É o conjunto de equações que fornece o comportamento previsto de um sistema de processos;
Neste conjunto inclui os dados de entrada necessários para a solução das equações;

Esses modelos são baseados em relações físicas e químicas conhecidas que incluem:
- Conservação de massa e energia;
- Reações cinéticas;
- Fenômenos de transporte;
- Relações termodinâmicas (equilíbrio de fase etc).

#### Modelagem de Processos:
A modelagem de processos do mundo real seguem passos muito bem definidos, sendo eles:
- 1°: Definição do problema (hipóteses).
- 2°: Mecanismos.
- 3°: Dados para o problema.
- 4°: Construção do modelo.
- 5°: Solução do modelo.
- 6°: Verifição da solução.
- 7°: Validação do modelo.




##### 1°: Definição do problema (hipóteses).
Nesta etapa devemos:
- Definir nosso sistema.
- Estabelecer hipóteses.
- Estabelecer os objetivos.
- Definir variáveis e inputs.

##### 2°: Mecanismos.
Nesta etapa devemos listar os fenômenos físicos-químicos mais importantes no nosso problema, por exemplo:
- Reação química
- Difusão de massa
- Condução de calor
- Transferência de calor
- Mudanças de fase
- Radiação
- Evaporação

##### 3°: Dados para o problema.
Nesta etapa devemos avaliar as informações que possuimos sobre o problema, como:
- Dados físico-químicos;
- Dados da planta;
- Parâmetros do equipamento;
- Cinética da reação.
    - Importante!! Caso os dados e valores do parâmetros não forem adequados deve-se voltar às etapas 1 e 2 para mudálos.
    
##### 4°: Construção do modelo.
Aqui definimos:
- Definir as hipóteses, volume de contorno e controle;
- Equações de constitutivas, taxa de reação, taxas de transferência etc;
- Condições iniciais e de contorno;
- Parâmetros.

##### 5°: Solução do modelo.
Nesta etapa executamos o modelo e asseguramos que o número de variáveis e graus de liberdade sejam satisfeitos.
Ademais, tais modelos podem ser compostos por:
- Sistemas algébricos;
- Equações diferenciais ordinárias – EDO;
- Equações diferenciais parciais – EDP;
- Equações algébrico-diferenciais – EAD.

##### 6°: Verifição da solução e  7°: Validação do modelo.
A verificação e validação do modelo são etapas que conferimos nossos resultados e comparamos com a realidade, calculamos o erro.
- 6°: Se o for verificado que o modelo não se comporta corretamente deve-se voltar ao passo 4 e 5.
- 7°: Se a validação resulta em que o modelo não representa o sistema voltar ao passo 2.
