## Reações em série:


#### Para este tipo de reação é muito importante o tempo de residência no reator. Se uma reação **A -> B é rápida** e **B -> C é lenta**, B pode ser maximizado por um tempo de residência menor. Mas, se a reação **A -> B é lenta** e **B -> C é rápida** será difícil maximizar B, pois, um <\tau> longo, maximizará o C.

---

<h3 align="center" style="color:gray">Efeito de pressão no cálculo de reatores</h3>

---

#### Considerar o efeito de pressão, quando:
- Toda vez que tiver o envolvimento do PBR.
- Quando tem o envolvimento de sólido e gás.

  Para reações em fase líquida, a concentração é muito pouco afetada pela pressão total do sistema. Assim, para a fase líquida pode-se ignorar os efeitos de queda de pressão quando estivermos dimensionando um reator.
    
  No entanto, em reações em fase **gasosa**, a concentração dos reagentes e produtos é proporcional á pressão total e, portanto, as vezes é preciso considerar a queda de pressão ao longo de um reator.
