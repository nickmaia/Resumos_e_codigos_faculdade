#### REPRESENTAÇÃO GRÁFICALINEAR DIRETA

Neste método, cada par de valores experimentais (Si ,vi ) origina uma reta num gráfico que representa Vmax em função de Km.

- Para o caso em que **S2 ≠ S1** e **v2 ≠ v1**, estas duas retas definem os mesmos pares de valores de Km e Vmax no ponto de intersecção. 
- Para o caso em que **v1 / S1 = v2 / S2**, casos em que as **duas retas são paralelas**, por não haver ponto de intersecção entre elas.

### Na Prática:
##### As melhores estimativas dos parâmetros Km e Vmax são dadas por:
- **Km:** mediana dos valores de Km i,j
- **Vmax:** mediana dos valores de Vmax i,j

Este método representa cada observação por uma reta e porque fornece diretamente os valores de Vmax e Km, assim é, a representação gráfica linear e a sua leitura direta.
