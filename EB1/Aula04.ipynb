<h5 align="center" style="color:black">Lei de Michaelis e Menten - Comportamento cinético</h5>

---

O modelo de Michaelis e Menten, explica a influência das concentrações iniciais de enzima e de substrato na velocidade inicial da reação enzimática.

- [EP] não existe e o produto não reverte para o substrato inicial.
- A taxa deve ser medida em um tempo pequeno o suficiente para que [S] permaneça praticamente constante.
- Para isto, somente 5% do substrato deve ser utilizado durante o tempo do ensaio.



---

<h5 align="center" style="color:black">Lei de Michaelis e Menten - Hipóteses</h5>

---

- O substrato e a enzima reagem reversivelmente entre si formando um ``composto intermediário`` complexo enzima-substrato ``(ES)``.
- O complexo formado se decompõe, regenerando a enzima e formando os produtos da reação.

<h5 align="center" style="color:black">Significado dos Valores de Km</h5>

---

- Usado para avaliar a afinidade entre a enzima e o substrato.
- Km é numericamente igual à concentração do substrato [mol/L].
- A concentração de substrato em que ``[S] = Km``, a equação de Michaelis-Menten torna-se ``V0 = Vmáx/2``.
- A concentração da enzima não afeta o Km.
- Valores baixos de Km refletem ``afinidade elevada`` da enzima pelo substrato e, portanto, esta atingirá a ``máxima eficiência catalítica`` em baixas concentrações de substrato. 
- Valores elevados de Km refletem baixa afinidade da enzima pelo substrato.
- Km varia de 10-1 a 10-7 M
- Km também está relacionado às constantes de velocidade de cada etapa da reação;
