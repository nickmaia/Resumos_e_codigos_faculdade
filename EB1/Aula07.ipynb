<h2 align="center" style="color:gray"> Inibição Enzimática</h2>

---

#### **Inibidor:** substância que reduz a taxa de uma reação;

#### Inibidor afeta a atividade da enzima, como:
- Combinando-se com elas;
- Influenciando negativamente a ligação do substrato;
- Influenciando a velocidade de reação;

### Tipos:



---

<h2 align="center" style="color:gray">Inibição Irreversível</h2>

---

Chamados de inativadores, ligam-se à enzima de maneira tão forte que bloqueiam permanentemente a atividade enzimática.
- Não há modelo cinético;
-  [EI] será acumulado;
-  [E] consumida (toda E se liga a I);
- A opção para este tipo de processo é eliminar o inibidor ou usar outra enzima que não sofre inibição

<h3 align="left" style="color:gray">Hipótese do Pseudo-estado estacionário (Hipótese de Briggs e Haldane)</h3>

---

#### Aumento aparente na constante Km, ou seja, um decréscimo aparente da afinidade da enzima pelo substrato sem afetar a velocidade máxima (Vmax).



#### O grau de inibição depende da relação [S]/[I]:
- Quanto maior for o [S] a taxa de velocidade aumenta e o grau de inibição caí.
- Quanto maior for o [I] o grau de inibição aumenta e a taxa de velocidade decresce.

---

<h3 align="left" style="color:black">Inibição não Competitiva</h3>

---

- O inibidor se liga á [E] em um ponto diferente do sítio ativo;
- Forma dois tipos de complexos:
  - [EI] e [EIS]
- Combinam com grupos funcionais da enzima, exceto com os sitios ativos.

- **Vmax** decresce, **Km** permanece inalterado;
- O grau de inibição depende de [I] e KI;
- atividade relativa, **a = Vi/V**;
- Para **[I] = KI**, então temos 50% de inibição para qualquer [S]

1. sem inibidor
2. com inibidor na concentração [I1]
3. com inibidor na concentração [I2] > [I1]

---

<h3 align="left" style="color:black">Inibição não Competitiva</h3>

---

Neste acaso o inibidor liga-se ao complexo ES, mas não à enzima livre.
Com isso, irá gerar um complexo inativo enzima-substrato-inibidor.

#### Tanto Vmax quanto Km são afetados pelo mesmo fator:




São observados decréscimos aparentes em Vmax e Km. 
O aumento aparente na afinidade da enzima por substrato e diminuição de Km é devido à ligação improdutiva do substrato, resultando em uma diminuição na concentração de enzima livre. Portanto, Vmax/2 será atingida com uma concentração de substrato relativamente menor.

1. sem inibidor.
2. com inibidor na concentração [I1]
3. com inibidor na concentração [I2] > [I1]
