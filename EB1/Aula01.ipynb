### Esquema geral de um bioprocesso


----

### BIOCATALISADORES

Um biocatalisador é um ``catalisador das reações bioquímicas dos seres vivos, são considerados biocatalisadores as enzimas, os hormônios e as vitaminas.`` Um biocatalisador atua na redução ou um aumento da energia de ativação de uma reação química, afetando a velocidade de uma reação, mas emergindo do processo inalterada.

### ENZIMAS
Enzimas são ``catalisadores biológicos responsáveis por aumentar a velocidade de uma determinada reação química.`` Sendo a maioria delas proteínas.

***OBS:*** Uma enzima geralmente catalisa uma só reação ou um conjunto de reações intimamente relacionadas.
