<h5 align="center" style="color:black">CARBOIDRATOS</h5>

----

##### CARACTERÍSTICAS

- Fonte de energia abundante e econômica para o homem, Também conhecidos como açúcares;
- Alguns são fonte de fibras dietéticas (celulose e hemicelulose);
- São compostos orgânicos com pelo menos 3 carbonos;
- fórmula química geral: ``Cn(H2O)n``
- Estrutura química dos carboidratos:    
    - Poliidroxialdeído
    - Poliidroxicetona


----

#### FUNÇÕES:


##### ENERGÉTICA


A principal função dos carboidratos é a ***_energética_***, sendo as fontes primárias de produção de energia sob a forma de ATP, cujas ligações ricas em energia são quebradas sempre que as células precisam de energia para as reações bioquímicas.

_ATP_: molécula de adenosina trifosfato, ao sofrer o processo de hidrólise (cisão por ação da água), essa molécula libera grande quantidade de energia livre.


##### ESTRUTURAL


A função estrutural é auxiliar na composição das estruturas celulares e também dos ácidos nucleicos, bem como libera a glicose assim que é ingerido e absorvido.

Exemplos:
- Parede celular dos vegetais (celulose);
- Carapaça dos insetos (quitina - exo-esqueleto);
- Células animais possuem uma série de carboidratos circundando a membrana plasmática que dão especificidade celular (glicocálix);

##### RESERVA ENERGÉTICA


Além de fornecer energia de maneira imediata, os carboidratos podem ser armazenados de diferentes formas. Nos vegetais, o carboidrato de reserva é o amido; nos animais, o carboidrato de reserva é o glicogênio.

***OBS***: Ambos são formados unicamente por glicose, unidas por ``ligações glicosídicas``.

##### SÍNTESE


Utilizado como fonte de átomos de carbono para a síntese de outros compostos celulares.

----

#### TIPOS:


Baseado no número de carbonos na estrutura química:
- 3C: triose
- 4C: tetrose
- 5C: pentose
- 6C: hexose

Baseado no número de unidades de açúcar na cadeia: (pesomolecular)
- Monossacarídeos: 1 unidade – glicose, frutose, galactose
- Dissacarídeos: 2 unidades – sacarose, maltose e lactose
- Oligossacarídios: 3 a 10 unidades – rafinose e estaquiose
- Polissacarídeos: mais de 10 unidades – amido, celulose, glicogênio

----

#### MONOSSACARÍDEOS

- Possuem uma única unidade de pollidroxialdeído ou pollidroxicetona;
- Não podem ser hidrolisados a formas mais simples (possuem de 3 a 9 átomos de carbonos na cadeia).
    ##### Glicose:
    - hexose, conhecida como dextrose;
    - açúcar.
    ##### Frutose:
    - açúcar mais doce/usado como adoçante/frutas;
    ##### Galactose:
    - faz parte do açúcar do leite (lactose)
    ##### Ribose:
    - parte da estrutura do DNA

#### MUTARROTAÇÃO

Mutarrotação se refere à variação na rotação ótica que
acompanha o equilíbrio de α e β

Os monossacarídeos com mais de cinco átomos de carbono, quando dissolvidos em água, apresentam o
fenômeno da _mutarrotação_, que consiste na mudança gradativa da rotação óptica até alcançar um ângulo de
equilíbrio. Esse comportamento se deve ao fato de que esses compostos apresentam duas estruturas
espaciais possíveis, uma das quais se transforma na outra durante a dissolução, até que um estado de
equilíbrio, onde ambas as estruturas coexistem, seja alcançado.

----

#### LIGAÇÃO GLICOSÍDICA

A ligação é chamada glicosídica se faz entre a
união de dois monossacarídeos. Esta ligação
covalente é formada com a perda de um
átomo de hidrogênio de um dos
monossacarídeos e a saída de um radical
hidroxila
hidrogênio
do outro. Com isto, forma-se uma
molécula de água.


***OLIGOSSACARÍDEOS:*** são constituídos de cadeias contendo unidades de monossacarídeos (variam de 2 até 10 unidades) unidas entre si por ligações glicosídicas.

    - Sacarose: (glicose e frutose) - açúcar da cana e da beterraba;
    - Lactose: (glicose e galactose) - açúcar do leite;
    - Maltose: (glicose e glicose) - presente nos grãos em germinação;
    

    
    
***DISSACARÍDEOS:*** São formados por duas moléculas de monossacarídeos ligados entre si por uma ligação glicosídica. São três os mais comumente encontrados nos alimentos, sendo constituídos por pelo menos uma molécula de glicose.

***POLISSACARÍDEOS:*** Polímeros de açúcares simples: (hexoses)

    Amido e glicogênio: união de várias moléculas de glicose
    


- ***CELULOSE:*** União de várias moléculas de glicose
    - faz parte da parede celular das plantas;
    - o funciona como fibra alimentar;

    


***AÇÚCARES REDUTORES:*** Açúcares capazes de reduzir os íons férricos ou cúpricos: o grupo carbonila (aldeído) é oxidado a grupo carboxila (ácido carboxílico);
- Glicose, frutose;

<h5 align="center" style="color:black">LIPÍDEOS</h5>

---

Os lipídios, também conhecidos como gorduras, são uma classe de moléculas biológicas formadas por ácidos graxos e álcool. 
- Compostos orgânicos heterogêneos insolúveis em água e solúveis em solventes orgânicos;
- Origem animal ou vegetal;

Principais funções:
- ``Estrutura das membranas celulares:`` cria uma barreira para a célula e controla fluxo de materiais;
- ``Armazenamento de energia:`` lipídeos são armazenados - tecido adiposo;
- ``Hormônios:`` estabelecem comunicações entre as células;
- ``Vitaminas:`` regulam processos biológicos;
- ``Isolantes Térmicos``;

São classificados em 2 grandes grupos:

***Simples*** - São compostos que por hidrólise dão origem somente a ácidos graxos e álcool; 

São divididos em:
- Óleos e gorduras - ésteres de ácidos graxos e glicerol – Acilgliceróis
- Ceras: ésteres de ácidos graxos e mono-hidroxiálcoois.

***Complexos*** - São compostos que apresentam outros grupos na molécula, além dos ácidos graxos e alcoóis;
- Ex.: Fosfolipídeos, Glicolipídeos, Esteroides;

##### ***ÁCIDOS GRAXOS*** - Ácidos monocarboxílicos de cadeias longas
Caracteristicas:
- entre 12 e 24 carbonos;
- número par de átomos de carbono;
- normalmente não ramificados;
- O ponto de fusão dos ácidos graxos saturados aumenta com o comprimento da cadeia carbônica:


#### Cadeias dos ácidos graxos:
***Saturadas:***
- Não possuem duplas ligações;
- São geralmente sólidos à temperatura ambiente;
- Gorduras de origem animal são geralmente ricas em ácidos graxos saturados (carne bovina, porco, galinha, gema do ovo... (principalmente produtos animais); óleo de coco, folhas de palmeiras);


***Insaturadas:***
- Possuem uma ou mais duplas ligações sendo mono (uma ligação dupla) ou poli-insaturados (duas ou mais ligações duplas);
- São geralmente líquidos à temperatura ambiente;
- Os óleos de origem vegetalsão ricos em Ácidos Graxos insaturados;
- Não são facilmente sintetizados pelos tecidos animais;
- Devem ser ingeridos através dos alimentos;

##### Saponificação:
Reação de um ácido graxo com NaOH.

<h5 align="center" style="color:black">PROTEÍNAS</h5>

---

#### Principais funções:
- ``Enzimas:`` atividade catalítica;
- ``Proteínas transportadoras:`` atuam como transportadoras de íons ou moléculas específicas.
    - Ex: hemoglobina - transporta o O2 dos pulmões até os tecidos periféricos;
-  ``Proteínas nutrientes:`` a ovoalbumina (principal proteína da clara do ovo); a caseína do leite;
-  ``Proteínas contráteis e de mobilidade:`` a actina e miosina dos músculos humanos e a dineína dos cílios e flagelos;
-  ``Proteínas estruturais:`` fornecem proteção ou resistência a estruturas biológicas. 
    - Ex: colágeno: couro, cartilagem; queratina: cabelo, unha;
-  ``Proteínas de defesa:`` as imunoglobulinas ou anticorpos, fibrinogênio e a trombina (coagulação sanguínea; prevenção de perda de sangue), venenos de serpentes e algumas toxinas bacterianas.
- ``Proteínas reguladoras:`` hormônios (insulina, regula o metabolismo de açúcares) e proteínas que se ligam ao DNA, regulando sua transcrição (leitura)

#### AMINOÁCIDOS

- 20 aminoácidos formam as proteínas;
    - Dos 20 aminoácidos que compõem as proteínas, o organismo não sintetiza 9. Estes são chamados de ``aminoácidos essenciais`` e devem ser obtidos através da alimentação. Os outros aminoácidos que o organismo sintetiza são chamados de
``aminoácidos naturais``;

- Todos são L-aminoácidos;
- Classificação de acordo com o radical (R):
    - não polar
    - polar não carregado
    - polar carregado positivamente
    - polar carregado negativamente

### Ligação Peptídica
- União entre dois aminoácidos formando um dipeptídio;
- União entre três aminoácidos através de duas ligações peptídicas formam um tripeptídio;
- União de mais de três aminoácidos formam um polipeptídio – Proteínas;



### CLASSIFICAÇÃO

- ***Composição:*** Baseada no número de cadeias polipeptídicas;
    - ``Monomérica:`` molécula de proteína formada por somente uma cadeia polipeptídica (mioglobina)
    - ``Oligomérica:`` molécula de proteína formada por duas ou mais cadeias polipeptídicas (hemoglobina)
- ***Tipo de moléculas:***
    - ``Simples:`` contém somente aminoácidos na cadeia
    - ``Conjugada:`` contém outras moléculas em sua estrutura (grupos prostéticos)
        - Ex: Glicoproteínas; Lipoproteínas; Metaloproteínas
        
        
***OBS:*** Os grupos prostéticos podem ser orgânicos ou inorgânicos e encontram-se ligados de forma firme à cadeia polipeptídica, muitas vezes através de ligações covalentes.

- ***Forma:***
    - ``Globular:``
        - Forma esférica;
        - Geralmente solúvel em água;
        - Desempenha funções dinâmicas (enzimas);       

    - ``Fibrosas:``
        - Forma alongada;
        - Geralmente insolúveis em água;
        - Desempenha funções estruturais (colágeno, queratina);
        



- ***DESNATURAÇÃO DAS PROTEÍNAS:***

``Desnaturação protéica`` resulta no desdobramento e na desorganização das estruturas secundárias e terciárias, sem que ocorra hidrólise das ligações peptídicas, ou seja, ocorra a quebra das ligações não covalentes.

- Conformação nativa (A) é destruída;
- Perda da atividade;
- Pode ser causada por:
    - Temperatura;
    - Alteração de pH;
    - Adição de solventes orgânicos polares;
    - Adição de detergentes e sabões;
