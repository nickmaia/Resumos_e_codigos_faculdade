<h3 align="center" style="color:gray">Momento de Força</h3>

---

### Conceito

---

Momento de uma força em relação a um ponto é o produto do
``módulo da força`` pela ``distância perpendicular`` do ponto a linha de
ação da força.

## <center>M = F x b </center>
- Unidades: N.m

---

### Sentido

---

#### Sentido dado pela regra da **Mão Direita**

#### O vetor **F** provoca rotação do corpo em torno do eixo O, deste modo a eficácia de F aumenta quando se aumenta o braço da força **b**.

### Exemplo:

#### Momento da força F em relação a um polo O



---

### Teorema de Varignon

---

O momento de uma força em relação a um polo é a soma dos momentos das suas componentes em relação a este mesmo polo.
