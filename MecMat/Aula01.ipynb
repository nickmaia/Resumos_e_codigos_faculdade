### Forças no Plano

---

- Força é uma grandeza vetorial. 
- **Módulo** (Intensidade): Caracterizada por um número de unidades - N(Newton). (Resultante)
- **direção:** Definida por sua linha de ação. (Ângulo)
- **sentido:** Caracterizado por uma seta.

1. Exemplo: sentido saindo do objeto, F1 = 10 [N]



2. Exemplo: sentindo entrando no objeto, F1 = -10 [N] 



---

### Efeitos:

---

- **Externos:** tende a modificar o estado de movimento original do corpo.



- **Internos:** tensão e deformação.



---
